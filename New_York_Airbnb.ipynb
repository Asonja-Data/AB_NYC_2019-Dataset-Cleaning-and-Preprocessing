```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for price
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["price"])
plt.title("Boxplot of Price")
plt.show()
```

```python
# Calculate the 99th percentile for price
price_cap = df["price"].quantile(0.99)
print(f"99th percentile price cap: {price_cap}")
```

```python
# Cap the prices at the 99th percentile
df["price"] = df["price"].clip(upper=price_cap)

# Confirm the new max price
print(f"New max price after winsorization: {df['price'].max()}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compare original vs. adjusted price distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df["price"], bins=50, kde=True)
plt.title("Price Distribution After Winsorization")

plt.subplot(1, 2, 2)
sns.boxplot(x=df["price"])
plt.title("Boxplot After Winsorization")

plt.show()
```
